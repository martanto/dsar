```python
from dsar import DSAR, PlotDsar
```

# Run DSAR

```python
dsar = DSAR(
    input_dir=r"D:\Data\OJN",
    start_date="2025-01-01",
    end_date="2025-01-08",
    station="OJN",
    channel="EHZ",
    network="VG",
    location="00",
    verbose=True,
    debug=True,
)
```

# Change the freq bands

```python
# "LF": [0.1, 4.5, 8.0]
dsar.first_bands(
    name="LF",
    first_freq=0.1,
    second_freq=4.5,
    third_freq=8,
)

# "HF": [0.1, 8.0, 16.0]
dsar.second_bands(
    name="HF",
    first_freq=0.1,
    second_freq=8.0,
    third_freq=16,
)
```

```python
dsar.run()
```

# Plot DSAR

```python
plot = PlotDsar(
    start_date="2024-01-01",
    end_date="2024-04-22",
    station="OJN",
    channel="EHZ",
    network="VG",
    location="00"
)
```

## Get combined dataframe to plot

```python
df = plot.df
```
